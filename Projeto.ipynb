# Quarto Objetivo


---

Análise para verificar a relação entre a quantidade de trancamentos e a quantidade de disciplinas ofertadas.

Obtendo dados sobre a quantidade de alunos que trancaram o curso ou tiveram perca de vínculo no BTI de 2013 até 2022.

```python
students_with_locked_status_df = pd.read_csv('https://raw.githubusercontent.com/JoaoVSouto/bti-data-analysis/main/locked-students-quantity-by-year/data/students-amount-by-semester.csv')
students_with_locked_status_df
```

Obtendo dados sobre a quantidade de disciplinas ofertadas no BTI (com e sem ênfase) de 2013 até 2022.

```python
bti_subjects_amount_df = pd.read_csv('https://raw.githubusercontent.com/JoaoVSouto/bti-data-analysis/main/offered-subjects/data/subjects-amount.csv')
bti_subjects_amount_df
```

```python
students_statuses_and_subjects_by_semester_df = pd.merge(students_with_locked_status_df, bti_subjects_amount_df, on='semester')
students_statuses_and_subjects_by_semester_df
```

Transformando o segundo semestre de cada ano no ponto médio entre os semestres adjacentes para que possa fazer sentido no gráfico de linhas que se sucede. A transformação ocorre da seguinte maneira `2020.2` -> `2020.6`.

```python
string_semesters = list(map(lambda x: str(x), students_statuses_and_subjects_by_semester_df['semester']))
parsed_semesters = list(map(lambda x: float(f'{x[:-1]}6' if x[-1] == '2' else x), string_semesters))
```

```python
plt.figure(figsize=(9,9))
plt.plot(parsed_semesters, students_statuses_and_subjects_by_semester_df['locked_students_amount'], label='Alunos com status trancado')
plt.plot(parsed_semesters, students_statuses_and_subjects_by_semester_df['canceled_students_amount'], label='Alunos com vínculo perdido')
plt.plot(parsed_semesters, students_statuses_and_subjects_by_semester_df['bti_subjects_all_emphasis'], label='Disciplinas ofertadas')
plt.plot(parsed_semesters, students_statuses_and_subjects_by_semester_df['bti_subjects_without_emphasis'], label='Disciplinas ofertadas (sem ênfases)')
plt.legend(loc='best')
plt.xticks(np.arange(min(parsed_semesters), max(parsed_semesters)+1, 1.0))
plt.xlabel('Semestres')
```

```python
odd_semesters = list(filter(lambda x: str(x)[-1] == '1', students_statuses_and_subjects_by_semester_df['semester']))
even_semesters = list(filter(lambda x: odd_semesters.count(x) == 0, students_statuses_and_subjects_by_semester_df['semester']))
print('Mean of students with locked situation by semester parity')
print('Odd semesters:', students_statuses_and_subjects_by_semester_df.query(f'semester in {odd_semesters}').loc[:, ['semester', 'locked_students_amount']]['locked_students_amount'].mean())
print('Even semesters:', students_statuses_and_subjects_by_semester_df.query(f'semester in {even_semesters}').loc[:, ['semester', 'locked_students_amount']]['locked_students_amount'].mean())
```

Calculando a correlação entre estudantes que trancaram o semestre com todas as disciplinas do BTI incluíndo todas as ênfases.

```python
students_statuses_and_subjects_by_semester_df.loc[:, ['locked_students_amount', 'bti_subjects_all_emphasis']].corr()
```

Calculando a correlação entre estudantes que trancaram o semestre com todas as disciplinas do BTI sem ênfase alguma.

```python
students_statuses_and_subjects_by_semester_df.loc[:, ['locked_students_amount', 'bti_subjects_without_emphasis']].corr()
```

```python
students_statuses_and_subjects_by_semester_df.rename(columns = { 'canceled_students_amount': 'Alunos c/ vínculo perdido', 'locked_students_amount': 'Alunos trancados', 'bti_subjects_all_emphasis': 'Matérias do BTI (todas ênfases)', 'bti_subjects_without_emphasis': 'Matérias do BTI (sem ênfase)' }, inplace=True)
corr_matrix = students_statuses_and_subjects_by_semester_df.drop('semester', axis=1).corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
